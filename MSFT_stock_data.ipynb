# Get stock market data for Microsoft (MSFT)

Import libraries

```python
import datetime as dt
import pandas as pd
from pandas_datareader import data as pdr
import yfinance as yf
import plotly.offline as py
import plotly.graph_objects as go
from plotly.subplots import make_subplots

yf.pdr_override()
py.init_notebook_mode(connected=True)
pd.options.plotting.backend = "plotly"
```

Get the dates we need for the project

```python
end_date = dt.datetime.now()
start_date = dt.datetime(2017, 1, 1)
start_date, end_date
```

Create a stock list

```python
stocklist = ['MSFT']
```

Create dataframe and fill it with stock data with pandas_datareader

```python
df = pdr.get_data_yahoo(stocklist, start=start_date, end=end_date)
df.head()
```

Create data subset with close data

```python
close = df.Close
close.head()
```

```python
close.describe()
```

Plot data subset 'close'. APPL vs MSFT.

```python
close.plot()
```

```python
close.pct_change().plot(kind='hist')
```
